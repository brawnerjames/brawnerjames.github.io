```python
import pandas as pd
```

```python
csv_file = "WebVisualizations/weatherdata.csv"
df = pd.read_csv(csv_file)
```

```python
df.set_index("Unnamed: 0")
df.rename(columns = {"lat":"Lat", "lon":"Lon", "city":"City", "clouds":"Clouds", "country":"Country", "humidity":"Humidity", "max_temp":"Max Temperature", "wind_sp": "Wind Speed"})
```

```python
data_html = df.to_html(buf=None, columns=None, col_space=None, header=True, index=True, na_rep='NaN', formatters=None, float_format=None, sparsify=None, index_names=True, justify=None, bold_rows=True, classes=None, escape=True, max_rows=None, max_cols=None, show_dimensions=False, notebook=False)
print(data_html)
```

```python
data_file = open("data.html", "w")
data_file.write(data_html)
data_file.close()
```
